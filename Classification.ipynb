```python
# clean the test set
hr_test = pd.read_csv('https://raw.githubusercontent.com/WHPAN0108/BHT-DataScience-S23/main/classification/data/Assigment/aug_test.csv')
```

```python
# change max and min values
hr_test = hr_test.replace({'experience': {'>20': 21}})
hr_test = hr_test.replace({'experience': {'<1': 1}})
hr_test = hr_test.replace({'last_new_job': {'>4': 5}})
hr_test = hr_test.replace({'last_new_job': {'never': 0}})
```

```python
# convert to numeric
hr_test['experience'] = pd.to_numeric(hr_test['experience'])
hr_test['last_new_job'] = pd.to_numeric(hr_test['last_new_job'])
```

```python
# If the column is numerical, impute the missing value as its median
mean_exp = hr_test['experience'].mean()
hr_test.loc[hr_test['experience'].isna(), 'experience'] = mean_exp
mean_job = hr_test['last_new_job'].mean()
hr_test.loc[hr_test['last_new_job'].isna(), 'last_new_job'] = mean_job
```

```python
# If the column is categorical, impute the missing value as its mode
mode_gen = hr_test['gender'].mode()
hr_test.loc[hr_test['gender'].isna(), 'gender'] = mode_gen
mode_uni = hr_test['enrolled_university'].mode()
hr_test.loc[hr_test['enrolled_university'].isna(), 'enrolled_university'] = mode_uni
mode_edu = hr_test['education_level'].mode()
hr_test.loc[hr_test['education_level'].isna(), 'education_level'] = mode_edu
mode_maj = hr_test['major_discipline'].mode()
hr_test.loc[hr_test['major_discipline'].isna(), 'major_discipline'] = mode_maj
mode_com = hr_test['company_type'].mode()
hr_test.loc[hr_test['company_type'].isna(), 'company_type'] = mode_com
```

```python
# convert categorical to numerical columns
cat_columns = ['gender','relevent_experience','enrolled_university','education_level','major_discipline','company_type']
hr_test2 = pd.get_dummies(hr_test, columns = cat_columns)
hr_test2
```

## Classification

```python
# split training data
x_train = hr_train2.drop(['target'], axis=1)
y_train = hr_train2['target']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(x_train)
```

```python
# testing different models
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

trees = 100
max_features = 3
results = []
names_of_models = []

model_list = [('LR', LogisticRegression()),
             ('KNN', KNeighborsClassifier()),
             ('DTC', DecisionTreeClassifier()),
             ('RFC', RandomForestClassifier(n_estimators=trees,max_features=3))]


for name, model in model_list:
    kfold = KFold(n_splits=10)
    cv_results = cross_val_score(model, scaled_X_train, y_train, cv=kfold, scoring='precision')
    results.append(cv_results)
    names_of_models.append(name)
    res = "{}: {} ({})".format(name, cv_results.mean(), cv_results.std())
    print(res)
```

```python
# Logistic Regression seems to have the highest accuracy, but could be better
from sklearn.model_selection import GridSearchCV
parameters_for_testing = {
"class_weight"    : ['balanced'],
 "multi_class"        : ['auto', 'ovr', 'multinomial']
}
model = LogisticRegression()

kfold = KFold(n_splits=10)
grid_cv = GridSearchCV(estimator=model, param_grid=parameters_for_testing, scoring='accuracy', cv=kfold)
result = grid_cv.fit(scaled_X_train, y_train)

print("Best: {} using {}".format(result.best_score_, result.best_params_))
means = result.cv_results_['mean_test_score']
stds = result.cv_results_['std_test_score']
params = result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("{}  {} with: {}" .format(mean, stdev, param))
```

```python
# Random Forest, just for comparison
parameters_for_testing = {
"n_estimators"    : [50,100] ,
 "max_features"        : [2,5],
}
model = RandomForestClassifier()

kfold = KFold(n_splits=5)
grid_cv = GridSearchCV(estimator=model, param_grid=parameters_for_testing, scoring='accuracy', cv=kfold)
result = grid_cv.fit(scaled_X_train, y_train)

print("Best: {} using {}".format(result.best_score_, result.best_params_))
means = result.cv_results_['mean_test_score']
stds = result.cv_results_['std_test_score']
params = result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("{}  {} with: {}" .format(mean, stdev, param))
```

```python
# using the tuned Random Forest
tuned_model_rf = RandomForestClassifier(n_estimators=100,max_features=5)
tuned_model_rf.fit(scaled_X_train, y_train)
predictions = tuned_model_rf.predict(scaled_X_train)
```

```python
# confusion matrix for train set
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_recall_curve,ConfusionMatrixDisplay
cm = confusion_matrix(y_train, predictions,labels=tuned_model_rf.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=tuned_model_rf.classes_)
disp.plot()
```

```python
# accuracy, precision, recall, and F1-score
print(classification_report(y_train,predictions))
```

```python
# split + scale test data
x_test = hr_test2.drop(['target'], axis=1)
y_test = hr_test2['target']
scaled_X_test = scaler.fit_transform(x_test)
```

```python
import pandas as pd
countries = pd.read_csv('https://raw.githubusercontent.com/WHPAN0108/BHT-DataScience-S23/main/python-DS/country.csv')
countries
```

```python
countries.describe()
```

```python
countries_eur = countries[countries['Currency'] == 'EUR']
countries_eur
```

```python
currencies = countries[['Name', 'Currency']]
currencies
```

```python
rich_countries = countries[countries['GDP'] > 2000]
rich_countries
```

```python
medium_countries = countries.query('Population > 50000000 and Population < 150000000')
medium_countries
```

```python
countries['GDP'].mean()
```

```python
countries['GDP'].fillna(0).mean()
```

```python
countries['Density'] = countries.Population/countries.Area
countries
```

```python
areas = countries.copy()
big = areas['Area']>1000000
small = areas['Area']<1000000
areas.loc[big, 'Area'] = 'BIG'
areas.loc[small, 'Area'] = 'SMALL'
areas
```
